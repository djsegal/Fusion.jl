```julia
Pkg.add("Tokamak")
Pkg.add("PyPlot")

using Tokamak
using PyPlot
```

```julia
Tokamak.load_input("input.jl", true)

T_list = 10:15

solved_eqs = Tokamak.sweep_T_k(T_list);
```

```julia
beta_solution = solved_eqs["beta"]
```

```julia
plot(beta_solution["B_0"],beta_solution["R_0"])
xlabel("B_0")
ylabel("R_0");
```

```julia
plot(T_list,beta_solution["B_0"],T_list,beta_solution["R_0"]);
xlabel("T")
legend(["B_0", "R_0"]);
```

```julia
limit_list = beta_solution["other_limits"]
```

```julia
plot(T_list,limit_list["beta"],T_list,limit_list["wall"]);
xlabel("T")
legend(["beta", "wall"]);
```
