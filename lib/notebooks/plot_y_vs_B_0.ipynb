```julia
sweep_options = [cur_key for cur_key in keys(first(values(solution_dict)))]

plot_options = ["custom"]

append!(plot_options, [cur_key for cur_key in keys(first(values(first(values(first(values(first(values(solution_dict)))))))))])

filter!(cur_option -> cur_option != "constraint", plot_options)

filter!(cur_option -> cur_option != "B_0", plot_options)

@manipulate for design in [keys(solution_dict)...], cur_plot in plot_options, sweep in sweep_options, custom_input="B_0^2 * R_0^3", yscale=["lin", "log"], xscale=["lin", "log"], transition=["show", "hide"], pinpoint=["show", "hide"]

    cur_design = design
    
    if cur_plot == "custom"
        depVar = custom_input
    else
        depVar = cur_plot
    end
 
    cur_regex = r"\b[a-zA-Z]\w*\b"
    all_matches = matchall(cur_regex, depVar)
    
    has_bad = ( parse( depVar ) == Expr(:incomplete, "incomplete: premature end of input") )
    
    if has_bad ; return ; end
    
    cur_plot_dict = OrderedDict()
    
    for (cur_key, cur_value) in solution_dict[cur_design][sweep]
        
        cur_plot_dict[cur_key] = OrderedDict(
            "x" => [], "y" => [], "transition" => nothing
        )
        
        initial_constraint = nothing
        
        for (cur_sub_key, cur_sub_value) in cur_value
            
            cur_anon = Module()
            has_nan = false

            for cur_matched_var in all_matches
                if !haskey(cur_sub_value, cur_matched_var)
                    has_bad |= true
                    break
                end

                cur_matched_value = cur_sub_value[cur_matched_var]

                if isnan(cur_matched_value)
                    has_nan = true
                    break
                end

                eval(cur_anon, parse( "$cur_matched_var = $cur_matched_value" )) 
            end

            if has_bad ; break ; end
            if has_nan ; continue ; end

            curXVar = cur_sub_value["B_0"]
            curYVar = eval(cur_anon, parse( depVar )) 

            if sweep == "new" && cur_key != "wave_theta = 150.0"
                continue
            end
            
            push!(cur_plot_dict[cur_key]["x"], curXVar)
            push!(cur_plot_dict[cur_key]["y"], curYVar)
            
            cur_constraint = cur_sub_value["constraint"]
            
            if cur_plot_dict[cur_key]["transition"] == nothing
                if initial_constraint == nothing
                    initial_constraint = cur_constraint
                elseif initial_constraint != cur_constraint
                    cur_transition_point = cur_plot_dict[cur_key]["x"][length(cur_plot_dict[cur_key]["x"])]
                    cur_transition_point -= cur_plot_dict[cur_key]["x"][length(cur_plot_dict[cur_key]["x"])-1]
                    cur_transition_point /= 2
                    
                    cur_transition_point += curXVar
                    
                    cur_plot_dict[cur_key]["transition"] = cur_transition_point
                end
            end
            
        end
        
        if has_bad ; break ; end
        
    end
    
    cur_layers = Array{Any}(length(cur_plot_dict))

    cur_legend_names = Array{Any}(length(cur_plot_dict))
    cur_legend_colors = map(x->"#$(hex(x))", distinguishable_colors(length(cur_plot_dict), lchoices = linspace(40, 70, 15), cchoices = linspace(30, 60, 15)))

    vline_width = ( transition == "show" ) ? 1.0mm : 0.0mm
    
    for (cur_index, (cur_key, cur_value)) in enumerate(cur_plot_dict)   
        if cur_index == 1
            cur_style= :dot
        elseif cur_index == 2
            cur_style= :dashdot
        elseif cur_index == 3
            cur_style= :dashdotdot
        else
            cur_style= :dash
        end
        
        if cur_value["transition"] == nothing
            cur_layers[cur_index] = layer(x=cur_value["x"], y=cur_value["y"], Geom.point, Geom.line, Theme(default_color=eval(parse("colorant\"$(cur_legend_colors[cur_index])\""))))
        else
            cur_layers[cur_index] = layer(x=cur_value["x"], y=cur_value["y"], xintercept=[cur_value["transition"]], Geom.point, Geom.line, Geom.vline(style=[cur_style],size=[vline_width]), Theme(default_color=eval(parse("colorant\"$(cur_legend_colors[cur_index])\""))))
        end
        
        cur_legend_names[cur_index] = cur_key
    end
    
    if ( pinpoint == "show")
        if cur_design == "act_1"
            pinpoint_data = act_1_data()        
        elseif cur_design == "act_2"
            pinpoint_data = act_2_data()
        else
            pinpoint_data = act_1_data()
#             error("No design data loaded")
        end
        
        x_pinpoint_layer = layer(xintercept=[pinpoint_data["B_0"]], Geom.vline(), Theme(default_color="grey"))
        push!(cur_layers, x_pinpoint_layer)
        
        try
            if depVar == "P_W" || depVar == "beta"
                cur_yintercept = 1.0
            else
                cur_yintercept = pinpoint_data[depVar]
            end
            
            y_pinpoint_layer = layer(yintercept=[cur_yintercept], Geom.hline(), Theme(default_color="grey"))
            push!(cur_layers, y_pinpoint_layer)            
        catch
            # whoopsie
        end
    end

    if !has_bad
        if yscale == "lin"
            if xscale == "lin"
                Gadfly.plot(
                    cur_layers...,
                    Guide.xlabel("B_0"), Guide.ylabel(""), Guide.title(depVar), 
                    Guide.xticks(ticks=B_list),
                    Guide.manual_color_key("Legend", cur_legend_names, cur_legend_colors)
                )  
            else
                Gadfly.plot(
                    cur_layers...,
                    Scale.x_log10,
                    Guide.xlabel("B_0"), Guide.ylabel(""), Guide.title(depVar), 
                    Guide.manual_color_key("Legend", cur_legend_names, cur_legend_colors)
                )  
            end            
        else
            if xscale == "lin"
                Gadfly.plot(
                    cur_layers...,
                    Scale.y_log10,
                    Guide.xlabel("B_0"), Guide.ylabel(""), Guide.title(depVar),
                    Guide.manual_color_key("Legend", cur_legend_names, cur_legend_colors)
                )  
            else
                Gadfly.plot(
                    cur_layers...,
                    Scale.x_log10, Scale.y_log10,
                    Guide.xlabel("B_0"), Guide.ylabel(""), Guide.title(depVar),
                    Guide.manual_color_key("Legend", cur_legend_names, cur_legend_colors)
                )  
            end                
        end
    end
    
end
```
